# Bifurcation diagram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 10))

b = np.linspace(0, 4, 399)
x = 0.5

for i in range(len(b)):
    x = b * x * (1 - x)
    plt.scatter(b, x, s=0.9, c='r')

plt.show()
```

```python
plt.figure(figsize=(15, 10))

def eq(b, x):
    return b * x * (1 - x)

def diagram(n_skip, n_iter, x0=0.5, step=0.0001):
    B = []
    X = []

    b_range = np.linspace(0, 4, int(1 / step))
    for b in b_range:
        x = x0
        for i in range(n_iter + n_skip + 1):
            if i >= n_skip:
                B.append(b)
                X.append(x)

            x = eq(b, x)

    plt.plot(B, X, ls='', marker=',')
    plt.xlabel('B')
    plt.ylabel('Xn+1')
    plt.show()

diagram(100, 10)
```
