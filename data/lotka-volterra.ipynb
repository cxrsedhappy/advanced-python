```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# a -> коэффициент прироста жертвы в отсутствии хищника
# b -> коэффициент прироста хищника в отсутствии жертвы
# d -> коэффициент смертности хищника
# g -> коэффициент истребления хищником жертв


def f(u, t, alpha, beta, delta, gamma):
    x, y = u
    return np.array([
        x * (alpha - beta * y),
        -y * (gamma - delta * x)
    ])
```

```python
t = np.linspace(0, 90, 1000)
u = [10, 10]
res = odeint(f, u, t, args=(1.1, 0.4, 0.1, 0.4))
x, y = res.T
```

```python
plt.figure()
plt.grid()
plt.plot(t, x, 'b', label='Rabbit')
plt.plot(t, y, 'r', label="Foxes")
plt.xlabel('Days')
plt.ylabel('Population')
plt.legend()

plt.show()
```

```python
plt.figure()
x = np.linspace(1.0, 3.0, 9) # initial conditions for rabbit population (prey)
for rabbit in x:
    u = [rabbit, 1.0]
    y = odeint(f, u, t, args=(1.1, 0.4, 0.1, 0.4))
    plt.plot(y[:,0], y[:,1], "-", label = "$x_0 =$"+str(u[0]))
plt.xlabel("Rabbit")
plt.ylabel("Foxes")
plt.legend()
```
